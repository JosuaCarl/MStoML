```python
def spectral_entropy(spectrum):
    return -np.sum(spectrum * np.where(spectrum != 0, np.log(spectrum), 0))

def spectra_entropy_divergence(spec1, spec2):
    return ( 2 * spectral_entropy(total_ion_count_normalization(spec1 + spec2)) -
             spectral_entropy(total_ion_count_normalization(spec1)) - spectral_entropy(total_ion_count_normalization(spec2)) 
            ) / np.log(4)
```
